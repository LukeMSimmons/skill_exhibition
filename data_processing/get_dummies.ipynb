## Get Dummy Variables | Cosmic Object Classification

We want to figure out a way to predict whether a celestial object is a galaxy, star, or quasar.. but first we're going to need some dummys

#### Import Libraries & Setup Engines

```python
from datetime import datetime as dt
import pandas as pd
import sqlalchemy as sql

sql_engine = sql.create_engine('mssql+pyodbc://@localhost')
```

#### Read SQL table
We are going to take a look at a dataset from the Sloan Digital Sky Survey that I've actually never seen before

```python
table = 'STAR_CLASSIFICATION'

query = f"SELECT * FROM {table}"
df = pd.read_sql_query(query, sql_engine, index_col='obj_ID')
df.head()
```

#### Use Pandas' get_dummies() function to get 3 dummy variables for each type of object
We want to keep the 'class' column for easier visualization later

```python
df.loc[:,'class_dum'] = df['class']
df = pd.get_dummies(df, columns=['class_dum'], prefix='', prefix_sep='')
df.head()
```

#### Let's write the table back to SQL for now
Check out the **feature_selection** notebook in the **explore_data** folder for more

```python
sql_engine.execute(f'DROP TABLE IF EXISTS {table}') 
df.to_sql(table, sql_engine) 
print('Completed: %s' % dt.now().strftime('%m/%d %H:%M\n'))
```
